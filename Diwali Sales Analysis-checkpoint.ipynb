```python
top_10=df['Product_Category'].value_counts().nlargest(5).index
filtered_df = df[df['Product_Category'].isin(top_10)]
filtered_df
```

```python
plt.figure(figsize=(20,6))
top_10=df['Product_Category'].value_counts().nlargest(10).index
filtered_df = df[df['Product_Category'].isin(top_10)]
ax=sns.countplot(x='Product_Category',data=filtered_df,order=top_10)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
asd=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
```

```python
plt.figure(figsize=(18,8))
ax=sns.barplot(x='Product_Category',y='Amount',data=asd)
for bars in ax.containers:
    ax.bar_label(bars)
```
